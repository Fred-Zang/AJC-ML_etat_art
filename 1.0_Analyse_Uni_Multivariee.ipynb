# mieux en 2 graphes sur le meme axe

```python
plt.figure(figsize=(12,8))

fig, ax = plt.subplots(2,1,sharex=True)
sns.histplot(data=df, x="tip", ax= ax[0])
sns.boxplot(data=df, x="tip", ax= ax[1])
```
